```python
import pandas as pd
import numpy as np
from google.colab import files
import csv
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
import statistics
import random
import datetime
import time
```

```python
## For the program to work as presented here, run on Google Colab

file_name=files.upload()
name='LEI_deltas_stages.csv'
```

```python
## Reading the file in 
df=pd.read_csv(name, sep=',')
rows=df.shape[0] #number of rows
columns=df.shape[1]  #number of columns
print(rows)
print (columns)
df.head()
```

```python
### Plotting the first 200 observations
fig, ax = plt.subplots(figsize=(15,10))
x = df['month'][0:200] 
y = df['CEI'][0:200]
ax.plot(x, y, color='black')
w=ax.xaxis.set_major_locator(plt.MaxNLocator(20))
plt.xticks(w, rotation=45)
plt.xlabel('Month')
plt.ylabel('CEI')
plt.title('CEI vs Stages of Economic Cycle Stage', size=14)
sns.despine()
 
for i in range(1,200,1):
 
  if df['color'][i]  =="dark green":
    ax.axvspan(i-1, i, color='green', alpha=0.45, linewidth=0 )
  elif df['color'][i]  =="light green":
    ax.axvspan(i-1, i, color='green', alpha=0.15, linewidth=0 )
  elif df['color'][i]  =="red":
    ax.axvspan(i-1, i,   color='red', alpha=0.15, linewidth=0 )
  elif df['color'][i]  =="orange":
    ax.axvspan(i-1, i, color='orange', alpha=0.15, linewidth=0 )
  else:
    pass
plt.savefig("CEI-200.png")
plt.show();
```
