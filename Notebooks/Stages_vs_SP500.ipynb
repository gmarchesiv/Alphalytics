```python
!pip install xlsxwriter
import pandas as pd
import numpy as np
from google.colab import files
import csv
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
import statistics
import xlsxwriter
import random
import datetime
import time
```

```python
file_name=files.upload()
name='SP_500_1985.csv'
```

```python
## Reading the file in 
df=pd.read_csv(name, sep=';')
rows=df.shape[0] #number of rows
columns=df.shape[1]  #number of columns
print(rows)
print (columns)
df.head()
```

```python
### Plotting From 1985 to 1991
fig, ax = plt.subplots(figsize=(15,10))
x = df['month'] [0:84]
y = df['close'] [0:84]
ax.plot(x, y, color='black')
w=ax.xaxis.set_major_locator(plt.MaxNLocator(9))
plt.xticks(w, rotation=45)
plt.xlabel('Month')
plt.ylabel('S&P 500')
plt.title('S&P 500 vs Economic Cycle Stage Jan 1985 - Dic 1991', size=14)
sns.despine()
 
for i in range(1,84,1):
  # print(i)
 
  if df['color'][i]  =="dark green":
    ax.axvspan(i-1, i, color='green', alpha=0.45, linewidth=0 )
  elif df['color'][i]  =="light green":
    ax.axvspan(i-1, i, color='green', alpha=0.15, linewidth=0 )
  elif df['color'][i]  =="red":
    ax.axvspan(i-1, i,   color='red', alpha=0.15, linewidth=0 )
  elif df['color'][i]  =="orange":
    ax.axvspan(i-1, i, color='orange', alpha=0.15, linewidth=0 )
  else:
    pass
plt.savefig("S&P 85-91.png")
plt.show();
```
